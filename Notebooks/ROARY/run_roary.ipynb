# Run Roary to generate a core gene alignment

At this stage you should have three GFF files generated by Prokka, each in its own directory. Provided your QC looked alright, you are now ready to run Roary to generate the pangenome. 

We are going to run Roary twice, first with the default settings, and then using MAFFT to generate a core alignment. For both of these runs we will want all the annotation files in the same directory, so lets take a copy of them to our current directory:

```bash
cp annotated_1368*/*.gff .
```

## Run Roary with default settings
Running Roary with the default settings is very straightforward. All you need to do is run `roary *.gff` and it will run on all the GFF files in the current directory. We want to run Roary twice with different settings, so in order to keep track of our output files from each run we will also specify an output directory for Roary to put the files in. Give the following command a go:

```bash
roary -f run_roary_no_alignment *.gff
```

This will run for a minute or two.

We will have a closer look at the results in the next section, so for now let us just see that there are some output files in the directroy we asked Roary to create:

```bash
ls -l run_roary_no_alignment
```

## Run Roary with Mafft
In order to create a multiFASTA alignment of the core genes, we will now run Roary again, but this time with some more options.

|Option|Description|
|------|-----------|
|-e|Create a multiFASTA alignment of the core genes|
|--mafft|Use with -e to use MAFFT instead of PRANK|
|-p|Number of threads to use|
  
For all usage options, you can have a look at the [Roary website](https://sanger-pathogens.github.io/Roary/).

```bash
roary -f run_roary_alignment -e --mafft -p 8 *.gff
```

This will take a bit longer to run than the previous command, about 5 minutes. Once finished you should have a directory called run_roary_alignment containing the output files, this time including a core_gene_alignment.aln file as well. Just quickly check that this is the case and then we will head over to the [next section](results.ipynb) to have a closer look at the results.

```bash
ls -l run_roary_alignment
```
